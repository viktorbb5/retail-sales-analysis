```python
import pandas as pd   #importing the file


df = pd.read_excel(r"C:\Users\beles\OneDrive\Desktop\cv portfolio\SQL - Retail Sales Analysis_utf .xlsb", engine='pyxlsb')

data_array = df.to_numpy(dtype=str)
```

```python
print(df.columns)       # Show all column names
print(df.head())        # Show first 5 rows

```

```python
import numpy as np

df_clean = df.dropna() # droping missing rows

data_array = df_clean.to_numpy() # convert to numpy array

print("Shape:", data_array.shape)
print(data_array[:5])   # cheking the shape and previewing the first 5 rows
```

```python
total_sales = data_array[:, -1].astype(float) #selecting all the rows and converting them in to floating point numbers#

print("Total Sales:", np.sum(total_sales)) # the sum off the total sales

print("Avarage sales:", np.mean(total_sales)) # the avatage off the total sales 

print("Max sales:", np.max(total_sales)) # the maximum value off the total sales
```

```python
categories = data_array[:, 6] # extracting the 7th column from the data base

unique, counts = np.unique(categories, return_counts=True) # identifying the unique categories and counting them

category_summary = sorted(zip(unique, counts), key=lambda x: x[1], reverse=True) # combinating and sorting


print("Most common product categories:")
for cat, count in category_summary:
    print(f"{cat}: {count}")
```

```python
prices = data_array[:, 8].astype(float)  # 9th column is price

print("Price statistics:")
print("Min:", np.min(prices))
print("Max:", np.max(prices))
print("Average:", np.mean(prices))
print("Median:", np.median(prices))

```

```python
expensive = data_array[prices > 100] # outputing the prices that are more than 100 euro

print("products over 100 euro:")
print(expensive)
```

```python
dates = data_array[:, 0]  # Assuming first column is dates in 'YYYY-MM-DD' format

months = [date[:7] for date in dates]  # Extract 'YYYY-MM'

unique_months, month_counts = np.unique(months, return_counts=True)

print("Sales per month:")
for m, count in zip(unique_months, month_counts):
    print(f"{m}: {count}")

    
    
```
